# Import

```python
# %pip install pandas
import pandas as pd
```

# Load

```python
df = pd.read_csv("eurofxref.csv")
df_hist = pd.read_csv("eurofxref-hist.csv")
df_hist.head()
```

```python
df.head()
```

# Selecting only USD, SEK, GBP, and JPY

```python
data = ["USD", "SEK", "GBP", "JPY"]
df_curr = df_hist[data]
df_curr.head()
```

# Data Cleaning

```python
df_curr.fillna(0, inplace=True)
df_curr.isnull().sum()
```

# Calculating rate mean

```python
df_curr_means = df_curr.mean()
```

# Creating another dataframe for final values

```python
# https://stackoverflow.com/questions/46276637/melt-and-find-average-counts-in-a-pandas-dataframe
# https://stackoverflow.com/questions/24216425/adding-a-new-pandas-column-with-mapped-value-from-a-dictionary
df_final = df_curr.iloc[:, :-1].melt(var_name="Currency Code", value_name="Rate").sort_values("Currency Code").reset_index(drop=True)
df_final["Mean Rate"] = df_final["Currency Code"].map(df_curr_means)
df_final
```

# Loading to markdown

```python
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_markdown.html
df_final.to_markdown("exchange_rates.md", index=False)
```
